```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Upload data
df = pd.read_csv("data.csv")
```

```python
# Control
df.head()
```

```python
# Total Sales
df['Sales_Amount'].sum()
```

```python
# Sales by region
region_sales = df.groupby('Region')['Sales_Amount'].sum()
region_sales
```

```python
# Graph
region_sales.plot(kind='bar')
plt.title("Sales by Region")
plt.show()
```

```python
# Customer type
customer_sales = df.groupby('Customer_Type')['Sales_Amount'].sum()
customer_sales
```

```python
# Graph
customer_sales.plot(kind='bar')
plt.title("Sales by Customer Type")
plt.show()
```

```python
# Top category
top_category = df.groupby('Product_Category')['Sales_Amount'].sum().sort_values(ascending=False)
top_category
```
